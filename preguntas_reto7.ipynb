Problema: Modelado del ruido en el centro de Cali utilizando funciones trigonométricas

La alcaldía de Cali ha decidido estudiar la distribución del ruido en diferentes zonas del centro de la ciudad. Se ha observado que los niveles de ruido cambian a lo largo del día, alcanzando su punto máximo durante las horas pico de tráfico vehicular, y disminuyendo considerablemente en la noche cuando el flujo vehicular y la actividad comercial se reducen.

Para medir el ruido en la zona los investigadores hicieron mediciones en diferentes momentos del día en 5 puntos diferentes de la ciudad. Además, los investigadores encontraron que el ruido en cada zona se puede modelar utilizando una función sinusoidal que refleja las variaciones a lo largo del día.

El nivel de ruido en decibeles $(dB)$ en una zona específica del centro de Cali se ha modelado utilizando una función sinusoidal que refleja estas variaciones. A continuación, se define la función que describe el nivel de ruido $R(t)$ en función del tiempo tt (en horas), donde $t=0$ corresponde a la medianoche:

$$ R(t)=A+B\sin\Big(\frac{2\pi t}{T}-\phi\Big)$$


**Interpretación de la función**



- $R(t)$ es el nivel de ruido en decibelios (dB) en un tiempo $t$ (medido en horas).
- $B$ es la amplitud de la variación del ruido a lo largo del día (en dB).
- $T$ es el período del ciclo de ruido (24 horas).
- $\phi$ es el desfase, que depende de la hora en la que el ruido máximo ocurre.
