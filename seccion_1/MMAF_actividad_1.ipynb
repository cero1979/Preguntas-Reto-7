### 1.4.1 Cálculo de la población de ardillas rojas en un parque  

Un equipo de científicos estudia la evolución de la población de ardillas en un parque natural durante un periodo de cinco años. Para ello, las ardillas se clasifican en dos grupos según su edad: **juveniles** y **adultas**.  

Cada año, la población de ardillas sigue el siguiente ciclo en ese orden:  

1. **Reproducción**: Cada ardilla adulta del año anterior da lugar a **dos** nuevas ardillas juveniles.  
2. **Maduración**: La mitad (**50%**) de las ardillas juveniles del año anterior alcanzan la adultez.  
3. **Supervivencia**: No todas las ardillas sobreviven. Las **ardillas adultas** del año anterior tienen una tasa de supervivencia del **80%** para el siguiente año.  

Al inicio del estudio, la población de ardillas es la siguiente:  

- **100 ardillas juveniles**  
- **50 ardillas adultas**  

El objetivo es determinar cuántas ardillas habrá en el parque después de **cinco años**, aplicando estas reglas (en orden) en cada ciclo anual.

#### 1.4.1.1 Manos a la obra

1. **Comprender el problema.** En esta etapa se busca comprender bien el problema que se quiere resolver e identificar las variables, datos y reglas relevantes para solucionar el problema. Descríbelos.

Se tienen dos grupos en la poblacion. Juveniles (J) y Adultas (A)
Cada año sigue un orden de evolucion
1. Reproduccion :cada adulta genera 2 juveniles.
2. Maduracion: El 50% de las juveniles del año anterior se convierten en adultos.
3. El 80% de los adultos del año anterior sobrevive al siguiente año.

Las condiciones iniciales del problema son: (Año 0)
$J_0$=100 , $A_0$=50

Se desea conocer cuantas ardillas habra en total al final del quinto año

Identifiquemos variables y reglas :
$J_n$: Ardillas juveniles en el año $n$, $A_n$ :ardillas adultas en el año $n$.
$T_n$ : Total de poblacion de ardillas que corresponde a $T_n=J_n+A_n$

Como se pasa del año $n$ al año $n+1$?

$J_{n+1}=2\times A_n$

$A_{n+1}=(0.8\times A_n)+(0.5\times J_n)$

```python
# Valores iniciales (Año 0)
J_0= 100 # Juveniles
A_0 =50 # Adultas

#Total inicial
T_0= J_0+A_0
print("Año 0 -> J_0:",J_0,"A_0",A_0,",T_0:", T_0)
```

2. **Planear la solución.** En esta etapa deberías identificar los pasos necesarios para resolver el problema y seleccionar la mejor estrategia para hacerlo.



Estrategia General:
1. Calcular en orden reproduccion, maduracion y supervivencia para pasar de $(J_n,A_n)$ a $(J_{n+1}, A_{n+1})$
2. Repetir el ciclo para $n=0,1,2,3,4$ hasta llegar a $(J_5,A_5)$
3. Sumar para obtener $T_5=J_5+A_5$

```python
#esto es una simple descripcion en codigo de la estrategia planteada
#J_{n+1}=2*A_n
#A_{n+1}=(0.8* A_n)+(0.5* J_n)
#T_{n+1}= J_{n+1}+A_{n+1}
```

3. **Ejecutar el plan.** En esta etapa deberías ejecutar la estrategia identificada en el paso anterior usando herramientas matemáticas, posiblemente asistidas por software.


Año 0 al año 1
$J_1=2\times A_0$ ,$A_1=(0.8\times A_0)+(0.5\times J_0)$

```python
#Año 0 -> Año 1
J_1=2*A_0
A_1 =(0.8*A_0)+(0.5*J_0)
T_1= J_1+A_1
print("Año 1 -> J_1:",J_1,"A_1",A_1,",T_1:", T_1)
```

Año 1 al año 2
$J_2=2\times A_1 , A_2= (0.8\times A_1)+(0.5\times J_1)$

```python
#Año 1 -> Año 2
J_2=2*A_1
A_2 =(0.8*A_1)+(0.5*J_1)
T_2= J_2+A_2
print("Año 2 -> J_2:",J_2,"A_2",A_2,",T_2:", T_2)

#Año 2 -> Año 3
J_3=2*A_2
A_3 =(0.8*A_2)+(0.5*J_2)
T_3= J_3+A_3
print("Año 3 -> J_3:",J_3,"A_3",A_3,",T_3:", T_3)


#Año 3 -> Año 4
J_4=2*A_3
A_4 =(0.8*A_3)+(0.5*J_3)
T_4= J_4+A_4
print("Año 4 -> J_4:",J_4,"A_4",A_4,",T_4:", T_4)


#Año 4 -> Año 5
J_5=2*A_4
A_5 =(0.8*A_4)+(0.5*J_4)
T_5= J_5+A_5
print("Año 5 -> J_5:",J_5,"A_5",A_5,",T_5:", T_5)
```
